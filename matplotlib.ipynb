```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)
y3 = np.cos(x*2)
plt.plot(x,y ,label="sin(x)",marker="." ,linestyle="-")
plt.plot(x,y2,label="cos(x)",marker="*")#se puede seguir sumando data 
plt.plot(x,y3,label="cos(x*2)")
plt.legend()
plt.title("Grafica de sin cos cosx*2")
plt.grid()
plt.xlabel("values")
plt.ylabel("result f(x)")
plt.yticks(np.arange(-2, 2.5, 0.5)) # no toma el 2.5
plt.show()      #se puede se parar con show los lienzos
```
